```python
# Import Dependencies and CSV File
import pandas as pd

# restaurants_csv = pd.read_csv("../Resources/az_yelp_business.csv")
restaurants_csv = pd.read_csv("../Resources/az_yelp_restaurants.csv")
restaurants_csv.head()
```

```python
# Filter out for only restaurants
rest = restaurants_csv.loc[restaurants_csv['categories'].str.lower().str.contains("restaurant")].reset_index(drop=True)
# Remove paranthesis
rest['name'] = rest['name'].str.replace('"','')
rest['address'] = rest['address'].str.replace('"','')
# Drop categories NA
rest['categories'] = rest['categories'].dropna()
# Lowercase categories for easier use
rest['categories'] = rest['categories'].str.lower().astype(str)
```

```python
# Remove non-restaurants
notRest = 'marketing|clothing|shops|advertising|schools|lakes|shopping|printing|convenience stores|yelp|grocery|golf|cafeteria|food court|stadiums & arenas|supernatural readings|veterinarians|hotels|event planning & services|hospitals|home services|markets|spas|adult entertainment|trucks|cinema'
rests = rest.loc[rest['categories'].str.contains(notRest)==False]
```
